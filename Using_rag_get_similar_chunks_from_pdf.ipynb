```python
# Step 2: Upload a PDF file
from google.colab import files
uploaded = files.upload()
```

```python
# Step 3: Extract text from PDF
import fitz  # PyMuPDF
pdf_file = list(uploaded.keys())[0]
doc = fitz.open(pdf_file)
text = ""
for page in doc:
    text += page.get_text()

print("PDF Loaded. Characters:", len(text))
```

```python
# ✂️ Step 4: Chunk the text
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.documents import Document

splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
chunks = splitter.split_text(text)
docs = [Document(page_content=chunk,
                 metadata={"source": f"chunk_{i}"}) for i
        , chunk in enumerate(chunks)]

print("✅ Text split into", len(docs), "chunks.")
```

```python
pip install -U langchain-community
```

```python
# 🧠 Step 5: Embed with HuggingFace
from langchain.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# 🧊 Step 6: Store in ChromaDB
from langchain.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    documents=docs,
    embedding=embedding,
    persist_directory="./chroma_store"
)

print("✅ Vector store created.")
```

```python
# 🔍 Step 7: Ask a question
retriever = vectorstore.as_retriever()
query = "What is thread"
results = retriever.get_relevant_documents(query)
```

```python
# 📤 Step 8: Show relevant chunks
print("\n🔎 Top relevant chunks:\n")
for i, doc in enumerate(results):
    print(f"Chunk {i+1}: {doc.page_content}\n")
```
